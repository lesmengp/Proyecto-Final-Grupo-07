```python
from google.cloud import storage

# Crea un cliente de almacenamiento de Google Cloud
client = storage.Client()

# Nombre del bucket y archivo
bucket_name = 'gmy'
blob_name = 'Yelp/business.pkl'

# Obtiene el bucket y el blob
bucket = client.get_bucket(bucket_name)
blob = bucket.blob(blob_name)

# Descarga el archivo a tu local
blob.download_to_filename('business.pkl')

# Ahora puedes trabajar con el archivo local 'business.pkl' en VS Code
```

```python
import pandas as pd
import gcsfs

# Ruta al archivo PKL en el bucket
bucket_name = 'gmy'
pkl_file = 'business.pkl'
gcs_path = f'gs://{bucket_name}/Yelp/{pkl_file}'

# Cargar el archivo PKL en un DataFrame
with gcsfs.GCSFileSystem().open(gcs_path, 'rb') as f:
    data = pd.read_pickle(f)

# Crear un DataFrame a partir de los datos
df = pd.DataFrame(data)

# Ahora "df" contiene los datos del archivo PKL en forma de DataFrame
```
