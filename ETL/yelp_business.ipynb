```python
import pandas as pd
import gcsfs

# Ruta al archivo PKL en el bucket
bucket_name = 'gmy'
pkl_file = 'business.pkl'
gcs_path = f'gs://{bucket_name}/Yelp/{pkl_file}'

# Cargar el archivo PKL en un DataFrame
with gcsfs.GCSFileSystem().open(gcs_path, 'rb') as f:
    data = pd.read_pickle(f)

# Crear un DataFrame a partir de los datos
df = pd.DataFrame(data)

# Ahora "df" contiene los datos del archivo PKL en forma de DataFrame
```
